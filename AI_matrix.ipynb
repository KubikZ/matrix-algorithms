```python
counter_add = 0
counter_mul = 0
counter_sub = 0
```

```python
def clear_counters():
    global counter_add, counter_mul, counter_sub
    counter_add, counter_mul, counter_sub = 0, 0, 0
```

```python
def add(A, B):
  global counter_add
  n = len(A)
  C = [[None for _ in range(n)] for _ in range(n)]
  for i in range(n):
    for j in range(n):
      C[i][j] = A[i][j] + B[i][j]
      counter_add += 1
  return C
```

```python
def sub(A, B):
  global counter_sub
  n = len(A)
  C = [[None for _ in range(n)] for _ in range(n)]
  for i in range(n):
    for j in range(n):
      C[i][j] = A[i][j] - B[i][j]
      counter_sub += 1
  return C
```
