**Recursive inverse**

```python
from common import *
from functions import *
from multiplication import *
import time
import tracemalloc
benchmark_size = 50
step = 10
```

```python
add_counters = []
sub_counters = []
mul_counters = []
time_counters = []
mem_counters = []
```

```python
for n in range(1, benchmark_size + 1, step):
    print("Computing n =", n)

    A = randomize_matrix(n)

    A = np.array(A)

    start_time = time.time()
    C = LU_factorization(A, binet)
    end_time = time.time()
    
    time_counters.append(end_time - start_time)
    reset_counters()
```

```python
for n in range(1, benchmark_size + 1, step):
    common.reset_counters()
    print("Computing n =", n)

    A = randomize_matrix(n)

    A = np.array(A)

    C = LU_factorization(A, binet)
    
    add_counters.append(common.counter_add)
    mul_counters.append(common.counter_mul)
    sub_counters.append(common.counter_sub)
```

```python
memory_counters = []

for n in range(1, benchmark_size + 1, step):
    print("Computing n =", n)

    A = randomize_matrix(n)

    A = np.array(A)

    tracemalloc.start()
    C = LU_factorization(A, binet)
    
    current, peak = tracemalloc.get_traced_memory()
    tracemalloc.stop()
    
    mem_counters.append(peak / 10**6)
```

```python
plot_all_metrics(step, time_counters, add_counters, sub_counters, mul_counters, mem_counters, "Binet")
```
