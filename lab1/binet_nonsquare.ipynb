```python
def binet(A, B):
    def rec(A, B):
        n = A.shape[0]
        m = A.shape[1]
        p = B.shape[1]
        assert(A.shape[1] == B.shape[0])

        if n == 1 and p == 1 and m == 1:
            return np.array([[A[0][0] * B[0][0]]])

        max_dim = max(n, m, p)
        if max_dim == m:
            # Splitting A and B
            mid_m = m // 2
            Avert1 = A[:, :mid_m]
            Avert2 = A[:, mid_m:]
            Bhor1 = B[:mid_m, :]
            Bhor2 = B[mid_m:, :]
            C = np.zeros((n, p))
            C = rec(Avert1, Bhor1) + rec(Avert2, Bhor2)
            return C
        elif max_dim == n:
            # Splitting A vertically
            mid_A = n // 2
            Avert1 = A[:mid_A, :] 
            Avert2 = A[mid_A:, :]
            C = np.zeros((n, p))
            C[:mid_A, :] = rec(Avert1, B) 
            C[mid_A:, :] = rec(Avert2, B)
            return C
        elif max_dim == p:
            # Splitting B horizontally
            mid_B = p // 2
            Bhor1 = B[:, :mid_B] 
            Bhor2 = B[:, mid_B:]
            C = np.zeros((n, p))
            C[:, :mid_B] = rec(A, Bhor1) 
            C[:, mid_B:] = rec(A, Bhor2)
            return C
        
    C = rec(A, B).tolist()
    return C
```
